```python
from transformers import pipeline

emotion_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)



def get_emotion_code_similarity(text):
    results = emotion_classifier(text)[0] # Get the list of score dicts
    return results

# Example usage (using the new function)
texts = [
    "I love this movie!",              # Expected: joy -> 2
    "This is annoying...",             # Expected: anger -> 1
    "I'm so sad right now.",           # Expected: sadness -> 3
    "Wow, that came out of nowhere!",  # Expected: surprise -> 4
    "Just another normal day.",         # Expected: neutral -> 0
    "I'm really scared of spiders!",   # Model might predict 'fear' highest. We want the highest score among our *target* labels.
    "That's absolutely disgusting."     # Model might predict 'disgust' highest. We want the highest score among our *target* labels.
]

for t in texts:
    print(f"Text: {t}")
    results = get_emotion_code_similarity(t)
    for result in results:
        if result['label'] in ['joy', 'anger', 'sadness', 'surprise', 'neutral', 'fear', 'disgust']:
            print(f"{result['label']}: {result['score']:.4f}")
    print()
    
    

```
